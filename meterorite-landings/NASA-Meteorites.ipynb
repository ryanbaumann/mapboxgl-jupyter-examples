```python
import pandas as pd
from mapboxgl.utils import *
from mapboxgl.viz import *
import os, json
```

```python
os.listdir()
```

```python
df = pd.read_csv('Meteorite_Landings.csv', parse_dates=['year'])
df.head(5)
```

```python
df_to_geojson(df[['name', 'id', 'recclass', 'mass (g)', 'year', 'reclat', 'reclong']].fillna(0), filename='metorites.geojson', 
              lon='reclong', lat='reclat')
```

```python
token = os.getenv('MAPBOX_ACCESS_TOKEN')
```

```python
# Generate data breaks and color stops from colorBrewer
measure_color = 'mass (g)'
color_breaks = [round(df[measure_color].quantile(q=x*0.1), 2) for x in range(3,11)]
color_stops = create_color_stops(color_breaks, colors='YlOrRd')

viz = CircleViz('metorites.geojson', 
                          access_token=token,
                          color_property = "mass (g)",
                          color_stops = color_stops,
                          zoom=3,
                          opacity=0.9,
                          center = (-95, 40),
                          below_layer = 'waterway-label'
                        )
viz.show()
```
